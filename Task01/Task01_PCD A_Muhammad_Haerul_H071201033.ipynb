```python
import numpy as np
import matplotlib as mpl
import matplotlib.image as image
import matplotlib.pyplot as plt
%matplotlib inline
```

###**1.Dengan menggunakan cita mountain.bmp:**

```python
#a.Transformasi menjadi citra biner menggunakan thresholding dengan nilai T =100!
#read img
img = image.imread("mountain.bmp")
m = img.shape[0] 
n = img.shape[1]
binary_img = np.zeros((m, n))
T = 100

#transform img to binary img
for i in range(m):
  for j in range(n):
    if (img[i, j] < T):
      binary_img[i, j] = 0
    else:
      binary_img[i, j] = 1

#show imgs
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap="gray")
ax[1].imshow(binary_img, cmap="gray")
plt.show()
```

```python
#b.Transformasi menjadi citra negatif!
img = image.imread("mountain.bmp") 
m = img.shape[0] 
n = img.shape[1] 
negative_img = np.zeros((m, n))

#transform img to negative img
for i in range(m):
  for j in range(n):
    negative_img[i, j] = 255 - img[i, j]
  
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap="gray")
ax[1].imshow(negative_img, cmap="gray")
plt.show()
```

```python
#c.Tingkatkan kecerahan citra dengan nilai b = 100!
img = image.imread("mountain.bmp") 
m = img.shape[0] 
n = img.shape[1] 
bright_img = np.zeros((m, n))
b = 100

#bright img
for i in range(m):
  for j in range(n):
    temp = img[i, j] + b
    #clipping
    if (temp < 0):
      bright_img[i, j] = 0
    elif (temp > 255):
      bright_img[i, j] = 255
    else:
      bright_img[i, j] = temp

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap="gray")
ax[1].imshow(bright_img, cmap="gray")
plt.show()
```

###**2. Dengan menggunakan citra girl-warna.bmp konversi citra menggunakan nilai rata-rata RGB!**

```python
img = image.imread("girl-warna.bmp") 

#extract the three channels (RGB) of img 
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

#vonvert img using the average of RGB
Y = (R + G + B)/3
converted_img = Y

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(converted_img, cmap="gray")
plt.show()
```

###**3. Kerjakan soal nomor 2 dengan menggunakan rumus Y = 0.299R + 0.587G + 0.144B!**

```python
img = image.imread("girl-warna.bmp") 

#extract the three channels (RGB) of the img
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

#convert the img using the formula
Y = 0.299*R + 0.587*G + 0.144*B
converted_img = Y

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(converted_img, cmap="gray")
plt.show()
```

###**4. Gabungkan citra bird.bmp(citra A) dan san.bmp (citra B) menggunakan rumus C = A + (1 - )B dimana nilai  = 0.8!**

```python
A = image.imread("bird.bmp")
B = image.imread("san.bmp")

#the operation can be executed if imgs have the same shape
if (A.shape == B.shape):
  m = A.shape[0]
  n = A.shape[1]
  C = np.zeros((m, n))
   = 0.8

  #merge imgs using the formula
  for i in range(m):
    for j in range(n):
       temp = *A[i, j] + (1 - )*B[i, j]
       if (temp < 0):
         C[i, j] = 0
       elif (temp > 255):
        C[i, j] = 255
       else:
        C[i, j] = temp
  
  fig, ax = plt.subplots(1, 3)
  ax[0].imshow(A, cmap="gray")
  ax[1].imshow(B, cmap="gray")
  ax[2].imshow(C, cmap="gray")
  plt.show()

else:
  print("The images don't have the same shape.")
  print("The operation can't be executed!")
```

###**5. Tunjukkan apakah terdapat perbedaan antara citra lena_gray.bmp dan lena_gray.gif!**

```python
bmp_img = image.imread("lena_gray.bmp")
gif_img = image.imread("lena_gray.gif")
m = bmp_img.shape[0]
n = bmp_img.shape[1]
d = 0

#show the difference of the img:
#shape diference
if (bmp_img.shape != gif_img.shape):
  d = d + 1
  print(d, ") The imgs have difference shape")
  print("bmp_img shape =", bmp_img.shape,"|", "gif_img shape =", gif_img.shape)

#pixels difference
gif_img = gif_img[:,:,0]
p = 0
for i in range(m):
  for j in range(n):
    if (bmp_img[i, j] != gif_img[i, j]):
      p = p + 1
if (p != 0):
  d = d + 1
  print(d, ") The imgs have difference value of the pixels by the number of =", p)

#have no difference
if (d == 0):
  print("The imgs don't have difference")

fig, ax = plt.subplots(1, 2)
ax[0].imshow(bmp_img, cmap="gray")
ax[1].imshow(gif_img, cmap="gray")
plt.show()
```

###**6. Tingkatkan kecerahan citra zelda.bmp dengan menggunakan perkalian dengan konstanta k = 2!**

```python
img = image.imread('zelda.bmp') 
m = img.shape[0] 
n = img.shape[1] 
bright_img = np.zeros((m, n))
k = 2

#bright img
for i in range(m):
  for j in range(n):
    temp = img[i, j] * k
    if (temp < 0):
      bright_img[i, j] = 0
    elif (temp > 255):
      bright_img[i, j] = 255
    else:
      bright_img[i, j] = temp

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img, cmap="gray")
ax[1].imshow(bright_img, cmap="gray")
plt.show()
```

###**7. Dengan menggunakan bird.bmp dan san.bmp tampilkan hasil operasi logika boolean or terhadap dua citra gray scale tersebut!**

```python
bird = image.imread("bird.bmp")
san = image.imread("san.bmp")

m = bird.shape[0]
n = bird.shape[1]
result = np.zeros((m, n))

#The operation of boolean logic or
for i in range (m):
  for j in range (n):
    result[i, j] = bird[i, j] or san[i, j]
  
fig, ax = plt.subplots(1, 3)
ax[0].imshow(bird, cmap="gray")
ax[1].imshow(san, cmap="gray")
ax[2].imshow(result, cmap="gray")
plt.show()
```

###**8. Rotasi citra lena_color.gif dengan sudut 30 dengan sudut putar titik tengah citra!**

```python
img = image.imread("lena_color.gif")
h, w, c = img.shape
angel = 30
rotated_img = np.zeros((h, w, c))

#convert rotation amount to radian
rot_rad = angel * np.pi / 180.0
rotate_m = np.array([[np.cos(rot_rad), np.sin(rot_rad)],
                     [- np.sin(rot_rad), np.cos(rot_rad)]])

#rotate and shift the indices
indices_org = np.array(np.meshgrid(np.arange(h), np.arange(w))).reshape(2, -1)
indices_new = indices_org.copy()
indices_new = np.dot(rotate_m, indices_new).astype(int)   
mu1 = np.mean(indices_new, axis=1).astype(int).reshape((-1, 1))
mu2 = np.mean(indices_org, axis=1).astype(int).reshape((-1, 1))
indices_new += (mu2-mu1)

#remove the pixels in the rotated image, that are now out of the bounds of the result image
t0, t1 = indices_new
t0 = (0 <= t0) & (t0 < h)
t1 = (0 <= t1) & (t1 < w)
valid = t0 & t1
indices_new = indices_new.T[valid].T
indices_org = indices_org.T[valid].T

xind, yind = indices_new
xi, yi = indices_org
rotated_img[xi, yi, :] = img[xind, yind, :]
rotated_img = rotated_img.astype(np.uint8)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(rotated_img)
plt.show()
```

###**9. Cerminkan citra lena_color.gif terhadap titik asal!**

```python
img = image.imread("lena_color.gif")
m = img.shape[0]
n = img.shape[1]
o = img.shape[2]
flip_img = np.zeros(((m, n, o)))
M = m-1
N = n-1

#flip img
for i in range(m):
  for j in range(n):
    for k in range (o):
      flip_img[i, j, k] = img[(N-i), (M-j), k]
flip_img = flip_img.astype(np.uint8)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(flip_img)
plt.show()
```

###**10. Perbesar citra girl-warna.bmp menjadi 512 x 512!**

```python
img = image.imread("girl-warna.bmp")
zoom_img = img
zoom_factor = 512/img.shape[0]

#increase the resolution of the img by repeating the pixels
for i in range(len(zoom_img.shape)-1):
    zoom_img = np.repeat(zoom_img, zoom_factor, axis=i)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img)
ax[1].imshow(zoom_img)
plt.show()
print("img size = ", img.shape, "|", "zoom_img size = ", zoom_img.shape)
```
