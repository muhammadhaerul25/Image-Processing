```python
import skimage
import cv2
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#read img
img = cv2.imread('cell darah.png')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
```

```python
#transform RGB to HSV
hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)

plt.imshow(hsv, cmap='gray')
```

### Note:
- So, we have to make a low and high color to detection the objects as main object

```python
#objects color
low_color = np.array([94, 80, 2])
high_color = np.array([156, 255, 255])

#color mask
object = cv2.inRange(hsv, low_color, high_color)

plt.imshow(object)
```

```python
#transform the img to binary img using threshold
ret, binary = cv.threshold(object, 100, 255, 0)

#find the img contours
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

#draw contours
result = cv2.drawContours(binary, contours, -1, (255,255,255), 3)

plt.imshow(result, cmap='gray')
```

```python
#show imgs
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(rgb, cmap='gray')
ax[0].set_title('Img Original')
ax[1].imshow(result, cmap='gray')
ax[1].set_title('Img Result')
```
